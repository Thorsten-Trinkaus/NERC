In order to execute every program related to the NLI approach, you firstly need to install each non-standard library from the **requirements.txt** file.

Then, to reproduce the data and results I have gathered during this project, run the following files in correct order.

---
1. **preprocessing_scripts.py**:
This program creates our pre-processed files for the NLI probing.

2. **nli_probing.ipynb**:
This file does the probing on our pre-processed files, the results being the predictions files for each dataset.

3. **nli_fine_tuning.ipynb**:
This file fine-tunes our model using the recently created prediction files.

4. **nli_probing_finetuned.py**:
This file uses our fine-tuned models we created from the step before and probes through our test set.\
**NOTICE**: When first opening the file, the code will be set to only run on the ontonotes files. To run the program for figer or ultra-fine simply change it in the "config" section at the top.

---
To recreate the evaluation and analysis steps, simply run:
- nli_evaluation.py

and for the analysis:
